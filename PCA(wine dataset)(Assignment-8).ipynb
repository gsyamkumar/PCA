```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
```

```python
#Reading the dataset
df = pd.read_csv('wine.csv')
df
```

# EDA

```python
df.info()
```

```python
#Converting the numerical data into z-scores
data_normal = scale(df)
data_normal
```

```python
pca = PCA(n_components=14)
pca_values =pca.fit_transform(data_normal)
pca_values
```

```python
#loading weights
pca.components_
```

```python
# The amount of variance that each PCA explains is 
#It explains how much information is present in the variance on each of pca values
 
var = pca.explained_variance_ratio_
var
```

```python
#Cumulative variance

#converting the pca values into percentages
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
#Variance plot for PCA components obtained
plt.plot(var,color= 'b')
```

```python
pca_values[:,0:1]
```

```python
final = pd.concat([pd.DataFrame(pca_values[:,0:2],columns=['pc1','pc2']),df[['Type']]],axis=1)
final
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
import seaborn as sns
sns.scatterplot(data=final,x='pc1',y='pc2',hue='Type',s = 100)
```

```python
p1 = sns.scatterplot(data=final,x='pc1',y='pc2',s = 100)  
for line in range(0,final.shape[0]):
     p1.text(final.pc1[line], final.pc2[line], final.Type[line], horizontalalignment='left', size='medium')
        
```

# Hierarchical Clustering

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('wine.csv')
df1
```

```python
# Normalization
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm = norm_func(df1.iloc[:,1:])
df_norm
```

```python
#Creating Dendogram using average method
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch 
z = linkage(df_norm, method="average",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,)
plt.show()     
```

```python
from sklearn.cluster import AgglomerativeClustering 
import warnings 
warnings.filterwarnings('ignore')
h_complete = AgglomerativeClustering(n_clusters=5, linkage='average',affinity = "euclidean").fit(df_norm) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
df1['clust']=cluster_labels # creating a  new column and assigning cluster_labels to new column 
df1
```

```python
df1.iloc[:,1:].groupby(df1.clust).mean()  
```

```python
#checking which data is following cluster1
data = df1[(df1.clust==1)]
data  
```

```python
#checking which data is following cluster2
data = df1[(df1.clust==2)]
data  
```

# K-means Clustering

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("wine.csv")
df
```

```python
#Normalization 
def norm_func(i):
    x =(i-i.min())/(i.max() -i.min())
    return (x)
```

```python
df3_norm= norm_func(df.iloc[:,1:])
df3_norm
```

```python
#Creating Elbow curve
fig = plt.figure(figsize=(10,8))
WCSS = []
for i in range(1,11):
    clf = KMeans(n_clusters=i)
    clf.fit(df3_norm)
    WCSS.append(clf.inertia_)#inerita is another name for WCSS and all the calculations are done in background
plt.plot(range(1,11),WCSS)
plt.title("Elbow Curve")
plt.xlabel('WCSS')
plt.ylabel("Number of Clusters")
plt.show()
```

```python
clf = KMeans(n_clusters=5)
y_kmeans = clf.fit_predict(df3_norm)  
```

```python
y_kmeans
```

```python
clf.cluster_centers_
```

```python
clf.inertia_
```

```python
md = pd.Series(y_kmeans)
df['clust'] = md
df
```

```python
df.iloc[:,1:].groupby(df.clust).mean() 
```

```python
df.plot(x="Alcohol",y ="Malic",c=clf.labels_,kind="scatter",s=50 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans')   
```
